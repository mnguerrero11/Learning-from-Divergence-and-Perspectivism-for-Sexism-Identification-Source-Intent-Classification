```python
import pandas as pd

# Rutas de los archivos CSV
input_path = '/content/drive/MyDrive/BackTranslation/Originales/spanish_all.csv'

# Leer el dataset original
df = pd.read_csv(input_path)

# Crear la columna 'label' con valores numéricos, 1 para 'sexist' y 0 para 'not sexist'
df['label'] = df['label_sexist'].apply(lambda x: 1 if x == 'sexist' else 0)

# Crear la columna 'value_task1' basado en la columna 'label'
df['value_task1'] = df['label'].apply(lambda x: 'YES' if x == 1 else 'NO')

# Asignar 'es' a la columna 'lang' para todos los registros
df['lang'] = 'es'

# Renombrar las columnas y seleccionar solo las columnas necesarias
df = df.rename(columns={'text': 'tweet'}).loc[:, ['tweet', 'label', 'value_task1', 'lang']]

# Agregar las otras columnas con valores predeterminados
df['test_case'] = 'EXIST2024'
df['split'] = 'TRAIN_ES'
df['id'] = df.index + 700000

# Seleccionar y reordenar las columnas en el orden deseado
df = df[['test_case', 'id', 'value_task1', 'lang', 'tweet', 'split']]
```

```python
# Contar el número de registros de cada clase en el dataset original
class_counts = df['value_task1'].value_counts()

print("Número de registros de cada clase:")
print(class_counts)
```

```python
output_path = '/content/drive/MyDrive/BackTranslation/Primitivos/es-backTr.csv'



# Guardar el nuevo dataset
df.to_csv(output_path, index=False)

print("Conversión completada. El nuevo dataset ha sido guardado en:", output_path)
```

```python
import pandas as pd

# Rutas de los archivos CSV
train_path = '/content/drive/MyDrive/data_task1_gold_primitivo/dataframes_task1_gold_primitivo_limpios/traducidos/en-valid.csv'
train_bcktr_path = '/content/drive/MyDrive/BackTranslation/Preprocesamiento y División/en_valid.csv'

# Leer los DataFrames desde las rutas
train_df = pd.read_csv(train_path)
train_bcktr_df = pd.read_csv(train_bcktr_path)

# Verificar las primeras filas de cada DataFrame
print("Primeras filas de train_df:")
print(train_df.head())

print("\nPrimeras filas de train_bcktr_df:")
print(train_bcktr_df.head())

# Concatenar train_df y train_bcktr_df por filas
concatenated_df = pd.concat([train_df, train_bcktr_df], axis=0)

# Mostrar las primeras filas del DataFrame concatenado
print("\nPrimeras filas del DataFrame concatenado:")
print(concatenated_df.head())

```
