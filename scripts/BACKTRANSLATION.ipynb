```python
from transformers import MarianMTModel, MarianTokenizer
import pandas as pd
from tqdm import tqdm
```

```python
def traducir_tweet(df, idioma_original, idioma_destino):
    # Cargar el modelo y el tokenizador para la traducción
    modelo = f'Helsinki-NLP/opus-mt-{idioma_original}-{idioma_destino}'
    tokenizer = MarianTokenizer.from_pretrained(modelo)
    modelo = MarianMTModel.from_pretrained(modelo)

    # Función para traducir un texto
    def traducir(texto):
        inputs = tokenizer(texto, return_tensors="pt", max_length=512, truncation=True)
        outputs = modelo.generate(**inputs)
        translated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
        return translated_text

    # Aplicar la función de traducción al DataFrame
    df_traducido = df.copy()  # Crear una copia del DataFrame original
    tqdm.pandas(desc="Traduciendo tweets")  # Inicializar la barra de progreso
    df_traducido['tweet'] = df_traducido['tweet'].progress_apply(traducir)  # Traducir el campo 'tweet'

    # Actualizar el campo 'lang' en el nuevo DataFrame
    df_traducido['lang'] = idioma_destino

    return df_traducido
```
