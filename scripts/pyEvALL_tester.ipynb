```python
import json

def concat_json(json_path1, json_path2):
    try:
        # Leer el contenido del primer archivo JSON
        with open(json_path1, 'r') as file1:
            data1 = json.load(file1)

        # Leer el contenido del segundo archivo JSON
        with open(json_path2, 'r') as file2:
            data2 = json.load(file2)

        # Concatenar las listas JSON
        concatenated_json = data1 + data2

        return concatenated_json

    except FileNotFoundError:
        print("No se pudo encontrar el archivo especificado.")
        return None
    except json.JSONDecodeError:
        print("Error al decodificar JSON. Asegúrate de que los archivos JSON sean válidos.")
        return None
```

```python
json_path1 = '/content/drive/MyDrive/predicciones/MODELO_BIN_1_3_DEBERTA_V3_BASE.json'
json_path2 = '/content/drive/MyDrive/predicciones/MODELO_BIN_1_4_ROBERTA_BASE_BNE.json'
resultado = concat_json(json_path1, json_path2)

resultado
```

```python
# Guardar el JSON concatenado en el archivo de salida
with open('/content/drive/MyDrive/predicciones/en-derberta_v3_base_es-roberta_base_bne.json', 'w') as output_file:
  json.dump(resultado, output_file, indent=4)
```

```python
from pyevall.evaluation import PyEvALLEvaluation
from pyevall.utils.utils import PyEvALLUtils
predictions = '/content/drive/MyDrive/predicciones/en-derberta_v3_base_es-roberta_base_bne.json'
gold = "/content/drive/MyDrive/EXIST2024_dev_task1_gold_hard.json"
test = PyEvALLEvaluation()
metrics=["Accuracy", "Precision", "Recall", "ICM", "ICMNorm" ,"FMeasure"]

params= dict()
params={PyEvALLUtils.PARAM_FORMAT: PyEvALLUtils.PARAM_OPTION_FORMAT_JSON }
report = test.evaluate(predictions, gold, metrics, **params)
report.print_report()
```
